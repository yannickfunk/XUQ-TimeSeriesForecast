```python
from pathlib import Path
import os
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
EXP_NAME = "sinus_peak_noise_valley_noise_conditional_lstm"
RESULTS_PATH = Path("../experiments") / EXP_NAME / "results"
POINT_TO_EXPLAIN = "scale_attributions_38.png"
```

```python
attr_methods = [e for e in os.listdir(RESULTS_PATH) if not e.endswith(".png")]
```

```python
for method in attr_methods:
    img = imread(RESULTS_PATH / method / POINT_TO_EXPLAIN)
    plt.imshow(img)
    plt.show()
```
